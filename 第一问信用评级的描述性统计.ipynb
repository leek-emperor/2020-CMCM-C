```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('违约风险.csv',encoding='gbk')
data
```

```python
A_max = data[data['信誉评级']=='A']['违约概率'].max()
A_min = data[data['信誉评级']=='A']['违约概率'].min()

B_max = data[data['信誉评级']=='B']['违约概率'].max()
B_min = data[data['信誉评级']=='B']['违约概率'].min()

C_max = data[data['信誉评级']=='C']['违约概率'].max()
C_min = data[data['信誉评级']=='C']['违约概率'].min()

D_max = data[data['信誉评级']=='D']['违约概率'].max()
D_min = data[data['信誉评级']=='D']['违约概率'].min()
```

```python
dd = {'A':[A_max,A_min],
     'B':[B_max,B_min],
     'C':[C_max,C_min],
     'D':[D_max,D_min]}

d1 = pd.DataFrame(data=dd,index=['Max','Min'])
d1
```

```python
A_des = data[data['信誉评级']=='A']['违约概率'].describe()
B_des = data[data['信誉评级']=='B']['违约概率'].describe()
C_des = data[data['信誉评级']=='C']['违约概率'].describe()
D_des = data[data['信誉评级']=='D']['违约概率'].describe()

dd2 = {'A':A_des,'B':B_des,'C':C_des,'D':D_des}
d2 = pd.DataFrame(data=dd2)
d2
```

```python
d2.to_csv('各类信誉评级的违约风险的描述性统计分析.csv',encoding='gbk')
```
